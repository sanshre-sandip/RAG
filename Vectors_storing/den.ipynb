```python
import os
import yt_dlp

playlist_url = input("Enter YouTube Playlist URL: ").strip()

if not playlist_url:
    print("Error: Playlist URL cannot be empty.")
    exit()

download_dir = os.path.expanduser("~/Videos/YouTube_Playlists")
os.makedirs(download_dir, exist_ok=True)

print(f"Videos will be saved to:\n{download_dir}\n")

ydl_opts = {
    "ignoreerrors": True,  # Skip unavailable/member-only videos
    "outtmpl": os.path.join(
        download_dir,
        "%(playlist_title)s",
        "%(title)s.%(ext)s"
    ),

    # Best video up to 1080p + best audio
    "format": "bestvideo[height<=1080]+bestaudio/best[height<=1080]",

    # Merge video and audio into mp4
    "merge_output_format": "mp4",
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([playlist_url])

print("\nDownload completed!")
```
