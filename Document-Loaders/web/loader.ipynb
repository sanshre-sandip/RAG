```python
from langchain_community.document_loaders import WebBaseLoader
```

```python
url = "https://www.nvidia.com/en-us/geforce/drivers/results/"
```

```python
loader = WebBaseLoader(url)
```

```python
docs = list(loader.lazy_load())
```

```python
print(docs[-1].page_content)
```
