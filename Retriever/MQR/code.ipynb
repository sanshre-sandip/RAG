```python
from langchain_community.vectorstores import FAISS
from langchain_core.documents import Document
from langchain_groq import ChatGroq
from langchain_classic.retrievers import MultiQueryRetriever
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
all_docs = [
    Document(page_content="Solar energy is a renewable source of energy obtained from sunlight. It helps reduce greenhouse gas emissions and lowers electricity costs."),
    Document(page_content="Wind energy is generated using wind turbines. It is a clean and sustainable energy source used worldwide."),
    Document(page_content="Hydropower generates electricity from flowing water. It is one of the oldest and most widely used renewable energy technologies."),
    Document(page_content="Fossil fuels such as coal, oil, and natural gas are non-renewable energy sources. Their combustion contributes to air pollution and climate change."),
    Document(page_content="Renewable energy includes solar, wind, hydropower, geothermal, and biomass energy. These sources can be replenished naturally and have lower environmental impacts."),
    Document(page_content="Artificial Intelligence (AI) enables computers to perform tasks that normally require human intelligence, such as language understanding, image recognition, and decision-making."),
    Document(page_content="Web development involves creating websites and web applications using technologies such as HTML, CSS, JavaScript, PHP, and databases."),
    Document(page_content="Regular physical exercise improves cardiovascular health, strengthens muscles, and helps maintain a healthy body weight."),
    Document(page_content="Online learning platforms allow students to access educational content from anywhere, making education more flexible and accessible."),
    Document(page_content="Football is one of the most popular sports in the world. It requires teamwork, strategy, and physical fitness."),
    Document(page_content="Newari culture is known for its rich traditions, festivals, architecture, cuisine, and the Nepal Bhasa language spoken by the Newar community."),

     
]
```

```python
%pip install sentence-transformers
```

```python
vector_store = FAISS.from_documents(
    documents=all_docs,
    embedding=HuggingFaceEmbeddings(),
    )
```

```python
similarity_retriever = vector_store.as_retriever(search_type="similarity", search_kwargs={"k": 5})
```

```python
query = "Benefits of solar and wind power"
```

```python
multiquery = similarity_retriever.invoke(query)
```
